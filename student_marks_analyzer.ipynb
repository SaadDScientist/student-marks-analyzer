```python
# creating empty dictionary for storing names and scores
data = {
    "Name" : [],
    "Marks" : [],
    "Status" : []
}

# fucntion calculating result status
def calculate_result(name, mark):
    # Determine Pass/Fail
    if mark >= 60:
        status = "Pass"
    else:
        status = "Fail"

        # Store data
    data["Name"].append(name)
    data["Marks"].append(mark)
    data["Status"].append(status)
```

```python
while True:
    name = input("Enter Student's Name: ")
    mark = int(input("Enter Student's Marks: "))

    calculate_result(name, mark)

        # Ask whether to continue
    user = input("Do you want to enter another student's marks? (y/n): ").lower()

    if user != "y":
        break

# Statistics calculation
average = sum(data["Marks"]) / len(data["Marks"])
highest = max(data["Marks"])
lowest = min(data["Marks"])

# Results
for i in range(len(data["Name"])):
    print(
        f"Name: {data['Name'][i]}, "
        f"Marks: {data['Marks'][i]}, "
        f"Status: {data['Status'][i]}"
    )

print("\nOverall Statistics")
print(f"Average Marks : {average:.2f}")
print(f"Highest Marks : {highest}")
print(f"Lowest Marks  : {lowest}")
```
